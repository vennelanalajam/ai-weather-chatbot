```python
import requests
import tkinter as tk
from tkinter import messagebox

API_KEY = "ffe15544082ca08e83466ff08a4df013"
BASE_URL = "http://api.openweathermap.org/data/2.5/weather?"

def get_weather():
    city = city_entry.get()
    if city.strip() == "":
        messagebox.showwarning("Input Error", "Please enter a city name")
        return
    
    url = BASE_URL + "q=" + city + "&appid=" + API_KEY + "&units=metric"
    response = requests.get(url)
    
    if response.status_code == 200:
        data = response.json()
        temp = data['main']['temp']
        weather = data['weather'][0]['description']
        humidity = data['main']['humidity']
        wind = data['wind']['speed']
        
        weather_icon = "üå§"
        if "rain" in weather.lower():
            weather_icon = "üåß"
        elif "cloud" in weather.lower():
            weather_icon = "‚òÅ"
        elif "clear" in weather.lower():
            weather_icon = "‚òÄ"
        
        result_label.config(
            text=f"{weather_icon} Weather in {city.title()}\n"
                 f"Condition: {weather.capitalize()}\n"
                 f"Temperature: {temp}¬∞C\n"
                 f"Humidity: {humidity}%\n"
                 f"Wind Speed: {wind} m/s",
            fg="white"
        )
    else:
        messagebox.showerror("Error", "City not found. Please try again.")

# GUI setup
root = tk.Tk()
root.title("üåç AI Weather Chatbot")
root.geometry("350x300")
root.configure(bg="#1E1E1E")  # dark background

# Heading
tk.Label(root, text="AI Weather Chatbot", font=("Arial", 16, "bold"), fg="cyan", bg="#1E1E1E").pack(pady=10)

# Input section
tk.Label(root, text="Enter City Name:", font=("Arial", 12), fg="white", bg="#1E1E1E").pack()
city_entry = tk.Entry(root, font=("Arial", 12))
city_entry.pack(pady=5)

# Button
tk.Button(root, text="Get Weather", command=get_weather, font=("Arial", 12, "bold"), bg="cyan", fg="black").pack(pady=5)

# Result label
result_label = tk.Label(root, text="", font=("Arial", 12), bg="#1E1E1E")
result_label.pack(pady=10)

# Footer
tk.Label(root, text="Powered by OpenWeatherMap API", font=("Arial", 8), fg="gray", bg="#1E1E1E").pack(side="bottom", pady=5)

root.mainloop()
```
